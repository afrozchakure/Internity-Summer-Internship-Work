```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ls 
```

```python
# Loading the dataset into a Data Frame
dataset = pd.read_csv('Salary_Data.csv')
```

```python
dataset.head() # Returns the first 5 elements of the Data Frame
```

```python
dataset.info()  # Provides info about our dataset
```

```python
# Converting our Data frame from pandas to numpy
# Assigning input and output
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1].values
```

```python
# Splitting our dataset into Train set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
# Importing LinearRegression class from sklearn library
from sklearn.linear_model import LinearRegression
model = LinearRegression()  # Creating an object of our class

# We have provided column as 1 but rows as unknown 
y_train = y_train.reshape(-1, 1)  
X_train = X_train.reshape(-1, 1)
X_test = X_test.reshape(-1, 1)

model.fit(X_train, y_train)
```

```python
# Predicting values
y_pred = model.predict(X_test)
y_pred
```

```python
# Calculating the performance metrics
from sklearn import metrics  
print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("RMSE", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

print("Score:", model.score(X_test, y_test))
```

```python
# Visualising the Training set results
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Visualising the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
